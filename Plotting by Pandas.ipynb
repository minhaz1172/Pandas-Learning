```python
import pandas as pd
data={"Month":["January","February","March","April","May"],
        "Sales":[100,250,560,300,400]}
df=pd.DataFrame(data)
print(df)
#plot line graph 
df.plot(x="Month",y="Sales",kind="line",title="Monthly Sales",legend=True,color="red")#plot line
df.plot(x="Month",y="Sales",kind="scatter",title="Monthly Sales",legend=True,color="red")
df.plot(x="Month",y="Sales",kind="pie",title="Monthly Sales",legend=True,color="red",autopct="%1.1f%%")
df.plot(x="Month",y="Sales",kind="bar",title="Monthly Sales",legend=True,color="red")
df.plot(x="Month",y="Sales",kind="barh",title="Monthly Sales",legend=True,color="red")
df.plot(x="Month",y="Sales",kind="box",title="Monthly Sales",legend=True,color="red")
df.plot(x="Month",y="Sales",kind="area",title="Monthly Sales",legend=True,color="red")
df.plot(x="Month",y="Sales",kind="hist",title="Monthly Sales",legend=True,color="red")
```
